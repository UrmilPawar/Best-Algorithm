```python
df1=df.fillna(method='ffill')
df2=df1.fillna(method='bfill')
df2
```

```python
df2.isnull().sum()
```

```python
from sklearn.preprocessing import OrdinalEncoder
enc=OrdinalEncoder()
enc.fit_transform(df2[['Location']])
df2[['Location']]=enc.fit_transform(df2[['Location']])
df2['Location']=df2['Location'].astype(float)
df2[['WindGustDir']]=enc.fit_transform(df2[['WindGustDir']])
df2['WindGustDir']=df2['WindGustDir'].astype(float)
df2[['WindDir9am']]=enc.fit_transform(df2[['WindDir9am']])
df2['WindDir9am']=df2['WindDir9am'].astype(float)
df2[['WindDir3pm']]=enc.fit_transform(df2[['WindDir3pm']])
df2['WindDir3pm']=df2['WindDir3pm'].astype(float)
df2[['RainToday']]=enc.fit_transform(df2[['RainToday']])
df2['RainToday']=df2['RainToday'].astype(float)
df2[['RainTomorrow']]=enc.fit_transform(df2[['RainTomorrow']])
df2['RainTomorrow']=df2['RainTomorrow'].astype(float)
df2.dtypes
```

```python
df2
```

```python
x=df2.iloc[:,1:22].values
y=df2['RainTomorrow'].values
print(x)
print(y)
```

```python
x=df2.iloc[:,1:22].values
y=df2['RainTomorrow'].values
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression(random_state=0).fit(x_train,y_train)
y_pred=clf.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(y_test,y_pred)
a=accuracy_score(y_test,y_pred)*100
a
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(x_train, y_train)
y_pred1=neigh.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(y_test,y_pred1)
a1=accuracy_score(y_test,y_pred1)*100
a1
```

```python
#SVM 
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(y_test,y_pred)
accuracy_score(y_test,y_pred)*100
```

```python
# Naive Bias
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(y_test,y_pred)
accuracy_score(y_test,y_pred)*100
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0)
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(y_test,y_pred)
accuracy_score(y_test,y_pred)*100

```

```python
# Random forest
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(x_train, y_train)
y_pred=clf.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(y_test,y_pred)
accuracy_score(y_test,y_pred)*100
```
